*Question 2.12: Quadrants*

```python
from math import atan2, degrees

x = float(input("Enter x: "))
y = float(input("Enter y: "))

a = degrees(atan2(y,x))

if x > 0 and y > 0:
        print("Angle is in quadrant 1.")

elif x > 0 and y < 0:
        print("Angle is in quadrant 4.")

elif x < 0 and y > 0:
        print ("Angle is in quadrant 2.")

elif x < 0 and y < 0:
        print ("Angle is in quadrant 3.")

else:
        print("Angle is not in one of the quadrants, but on one of the borders.")
```

What is special about atan2() compared to math.atan() is that atan2() takes two arguments, whilst atan() takes only one argument. Also,  atan2 returns the right angle, so it gives the right quadrant. Atan could give you an angle of 45 degrees, whilst the actual angle is for example 135 degrees.

*Question 2.15: Intersection of two lists*

```python
night1 = [1, 3, 2, 4, 0, 3, 7, 5, 2, 9, 2, 4, 3]
night2 = [0, 0, 2, 4, 2, 4, 3, 2, 1, 1, 1, 2, 0]

Night1 = set(night1)
print(f"Set version of night1: {Night1}")
Night2 = set(night2)
print(f"Set version of night2: {Night2}")

inters = set.intersection(Night1, Night2)
print(f"The intersection of Night1 and Night2 is: {inters}")

if 5 in inters:
        print(f"5 is found in the intersection of Night1 and Night2.")
else:
        print(f"5 is not found in the intersection of Night1 and Night2.")

if 5 in Night1:
    print(f"5 is a member of the set version of night1.")
else:
    print(f"5 is not a member of the set version of night1.")
```

*Question 2.18: Multiple references to the same sequence object*

```python
a = b = [1, 2, 3, 4]
c = b

a[2] = 0
print(a)
print(b)
print(c)
```

This gives 3 equal lists, where the 3 has changed into a 0. Lists can be changed, sets cannot be changed. You can copy it by using for example .copy(). If you then change a, only a will be changed.
